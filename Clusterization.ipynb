```python
from random import choice
from collections import defaultdict

import numpy as np
from scipy.spatial import distance
from scipy import stats


INFINITY = 10 ** 10


class KMeans:
  def __init__(self, n_centroids: int):
    self.data = []
    self.centroids = []
    self.n_centroids = n_centroids
  
  def _pick_init_centroid(self, data: list):
    return choice(data)
  
  def _get_point_dists_to_closest_centroid(self, data: list, centroids: list) -> dict:
    dists = defaultdict(lambda: INFINITY)
    d_sum = 0
    for i, point in data:
      for j, centroid in centroids:
        dists[i] = min(dists[i], distance.euclidean(point, centroid) ** 2)
      d_sum += dists[i]
    return {"d2_sum": d_sum, "dists": dists}
  
  def _define_new_centroid(self, data: list, centroids: list) -> int:
    dists = self._get_point_dists_to_closest_centroid(data, centroids)
    dists, cum_sum = dists["dists"], dists["d2_sum"]
    curr_cum_sum = 0
    curr_point = 0
    rand_k = stats.truncnorm.rvs(0, 1, loc=0.5, scale=0.3)
    while rand_k * cum_sum > curr_cum_sum:
      curr_cum_sum += dists[curr_point]
      curr_point += 1
    if curr_point < len(data):
      return data[curr_point]
    return data[-1]
  
  def _define_centroids(self, data: list, n_centroids: int):
    init_centroid = self._pick_init_centroid(data)
    centroids = [init_centroid]
    for _ in range(n_centroids - 1):
      centroids.append(self._define_new_centroid(data, centroids))
    return centroids
  
  def prepare(self, data: list, n_centroids: int = None):
    self.data = data
    data = list(enumerate(data))
    self.centroids = self._define_centroids(data, n_centroids or self.n_centroids)
  
  def get_closest_centroid(self, data_line: list):
    min_dist = INFINITY
    min_dist_el = None
    for centroid in self.centroids:
      centroid_coords = centroid[1]
      if min_dist > distance.euclidean(data_line, centroid_coords):
        min_dist = distance.euclidean(data_line, centroid_coords)
        min_dist_el = centroid
    return min_dist_el
  
  def get_classes(self, data_lines: list):
    centroids_to_classes = dict([(centroid[0], i) for i, centroid in enumerate(self.centroids)])
    classes = []
    for data_line in data_lines:
      closest_centroid = self.get_closest_centroid(data_line)
      classes.append(centroids_to_classes[closest_centroid[0]])
    return classes
```

```python
km = KMeans(3)

test_data = [[1, 2, 3], [1, 2, 3], [4, 5, 6], [6, 7, 5], [2, 2, 2], [100, 9, -9]]
km.prepare(test_data)
```

```python
from collections import Counter


class DBScan:
  def __init__(self, min_neighbours_n: int, min_d: int):
    self.min_neighbours_n = min_neighbours_n
    self.min_d = min_d
    self.buckets = None
    self.bp_to_cluster = None
    self.base_points = None
    self.edge_points_clusters = None
  
  def _get_closest(self, data: dict, point: list, min_d: int):
    point_coords = data[point]
    for i, coords in data.items():
      if i == point:
        continue
      if distance.euclidean(point_coords, coords) <= min_d:
        yield (i, coords)
  
  def _get_closest_base(self, data: dict, base_points: dict, point: list, min_d: int):
    point_coords = data[point]
    for neighbour in self._get_closest(data, point, min_d):
      if neighbour[0] in base_points:
        yield neighbour
  
  def _define_base_points(self, data: dict, min_neighbours_n: int, min_d: int):
    base_points = {}
    for i, coords in data.items():
      if len(list(self._get_closest(data, i, min_d))) >= min_neighbours_n:
        base_points[i] = coords
    return base_points
  
  def _define_edge_points_clusters(self, data: dict, base_points: dict, bp_to_cluster: dict, min_neighbours_n: int, min_d: int):
    edge_points_clusters = {}
    for i, coords in data.items():
      if i in base_points:
        continue
      closest_base_points = list(self._get_closest_base(data, base_points, i, min_d))
      clusters_counter = Counter([bp_to_cluster[tuple(point_i[1])] for point_i in closest_base_points])
      if len(closest_base_points) > 0:
        edge_points_clusters[tuple(coords)] = bp_to_cluster[tuple(sorted(sorted(closest_base_points, key=lambda el: distance.euclidean(el[1], coords)), key=lambda cl_el: clusters_counter[bp_to_cluster[tuple(cl_el[1])]])[-1][1])]
    return edge_points_clusters
  
  def _get_base_points_buckets(self, base_points: list, min_neighbours_n: int, min_d: int):
    buckets = []
    for i, coords in base_points.items():
      buckets.append([(i, coords)])
      deleted_inds = set()
      for b_i, bucket in enumerate(buckets):
        if b_i == len(buckets) - 1:
          continue
        for j, packed_coords in bucket:
          if distance.euclidean(coords, packed_coords) <= min_d and j != i:
            buckets[-1].extend(bucket)
            deleted_inds.add(b_i)
            break
      buckets = [b for b_i, b in enumerate(buckets) if b_i not in deleted_inds]
    return buckets

  def prepare(self, data: list, min_neighbours_n: int = None, min_d: int = None):
    self.data = data
    data = dict(enumerate(data))
    min_neighbours_n = min_neighbours_n or self.min_neighbours_n
    min_d = min_d or self.min_d
    self.base_points = self._define_base_points(data, min_neighbours_n, min_d)
    print(f'BPs: {self.base_points}')
    self.buckets = self._get_base_points_buckets(self.base_points, min_neighbours_n, min_d)
    print(f'Buckets: {self.buckets}')
    self.bp_to_cluster = dict([(tuple(bp[1]), i) for i, bucket in enumerate(self.buckets) for bp in bucket])
    print(f'BP2Cs: {self.bp_to_cluster}')
    self.ep_to_cluster = self._define_edge_points_clusters(data, self.base_points, self.bp_to_cluster, min_neighbours_n, min_d)
    print(f'EP2Cs: {self.ep_to_cluster}')

  def get_clusters(self, data: list):
    clusters = []
    for entry in data:
      entry = tuple(entry)
      if entry in self.bp_to_cluster:
        clusters.append(self.bp_to_cluster[entry])
      elif entry in self.ep_to_cluster:
        clusters.append(self.ep_to_cluster[entry])
      else:
        clusters.append(None)
    return clusters
```
